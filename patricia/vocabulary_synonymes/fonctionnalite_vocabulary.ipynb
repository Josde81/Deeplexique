# Vocabulary

```python
import pandas as pd
import numpy as np
```

```python
with open('LaParure.txt', encoding='latin-1') as f:
    text = f.read()
```

## Tokenization

```python
from gensim.utils import tokenize
```

```python
tokens = list(tokenize(text, lowercase=False))
```

## Liste Stop words

```python
import json
with open('stopwords-fr.json') as f:
    text_file = f.read()
    StopWord = json.loads(text_file)
    
```

```python
StopWord
```

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#nouveau text splité en phrases

text2 = text.replace("\n", "").split(".")
text2
```

```python
vectorizer = CountVectorizer(stop_words=StopWord)
X = vectorizer.fit_transform(text2)

frequence_mots = X.toarray()
```

```python
frequence_mots.shape
```

```python
compte = np.sum(frequence_mots, axis=0)
```

```python
#il associe le nombre d'occurence aux mots et il fait un DataFrame

compte = list(compte)
mots = vectorizer.get_feature_names()

mots_occurence = pd.DataFrame(data={'compte' : compte, 'mots': mots})
mots_occurence = mots_occurence.sort_values(by = 'compte', ascending=False)
mots_occurence.iloc[:30]
```

```python
#je fais un json avec les mots du text
json_str = mots_occurence.to_json(orient='records')
file = open('./test.json','w+')
file.write(json_str)
file.close()
```

## Dictionnaire des synonymes

```python
import json
```

```python
with open("./thesaurus-v2.3/dic_synonymes.json") as f:
    data = f.read()
```

```python
dic_synonymes = json.loads(data)
```

```python
dic_synonymes["aventuré"]
```

## Récupérer les synonymes

```python
mot_et_synonymes = {mot:dic_synonymes[mot] for mot in mots_occurence["mots"] if mot in dic_synonymes}
```

```python

def get_synonymes(mots):
    return {mot:dic_synonymes[mot] for mot in mots if mot in dic_synonymes}

get_synonymes("femme")
    
```

```python
list(mot_et_synonymes.items())[:20]
```
